```python
import random # 계좌개설을 위한 랜덤모듈 불러오기

class Account: # camellcase A로 대문자 시작
    count = 0     # 클래스 변수는 클래스 바로 아래에 위치. 모든 인스턴스 공유.
    def __init__(self, name, money): # name money 사용자 속성 설정
        self.bank = "SC은행" # 은행 이름 고정 ("" 변하지 않는다)
        self.name = name
        self.money = money
        self.deposit_count = 0
        self.deposit_list = []
        self.withdraw_list = []

                        # randint 정수로 출력
        part1 = str(random.randint(100, 999)) # 100 ~ 999
        part2 = str(random.randint(0, 99)).zfill(2) # zfill 문자열
        part3 = str(random.randint(1, 999999)).zfill(6)

        self.account_number = part1+"-"+part2+"-"+part3
        Account.count += 1

    def __str__(self):
      return f"은행이름: {self.bank}, 예금주: {self.name}"

    def get_account_num(self):
      return Account.count

    def deposit(self, amount):   # amount : 입금할 금
      if amount < 1:
        print("입금 불가")
      else:
        self.money += amount
        self.deposit_count += 1
        self.deposit_list.append(amount)
        if self.deposit_count == 5:
          self.money += self.money * 0.01


    def withdraw(self, amount):
      if amount > self.money:
        print("출금 불가")
      else:
        self.money -= amount
        self.withdraw_list.append(amount)

    def display_info(self):
      print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔액: {self.money}")

    def deposit_history(self):
        print(f"입금 내역: {self.deposit_list}")

    def withdraw_history(self):
        print(f"출금 내역: {self.withdraw_list}")


# class 끝

# 계좌 3개 만들기
acc1 = Account("홍길동", 10000)
acc2 = Account("김철수", 20000)
acc3 = Account("이영희", 1500000)

# 계좌 개수 출력
print(f"총 계좌 개수 : {acc1.get_account_num()}")

# 입금 테스트
acc1.deposit(5000)
acc1.display_info()

# 출금 테스트
acc1.withdraw(3000)
acc1.display_info()

# 잔고 100만원 이상 출력
# accounts = [acc1, acc2, acc3]
# for acc in accounts:
    # if acc.money >= 1000000:
      # acc.display_info()

# 디버깅 내용
# 백만원 이상 조건 만족 출력이 안되었음.
# 출력 코드 작성되지 않아 새로 추가하였음.
# 이후 출력 결과 만족

print("\n--- 잔고 100만원 이상 고객 검색 시작 ---")
accounts = [acc1, acc2, acc3]
for acc in accounts:
    print(f"검사 중: {acc.name}, 잔액: {acc.money}") # 루프가 돌고 있는지 확인
    if acc.money >= 1000000:
        print(">> 조건 만족! 정보 출력:")
        acc.display_info()
    # else:
        # print(">> 조건 미달")

# 입금/출금 내역 출력
acc1.deposit_history()
acc1.withdraw_history()




```

```python
import random

class Character:
    def __init__(self, name, hp, attack, defense):
        self.name = name
        self.hp = hp
        self.max_hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense)
        self.hp -= actual_damage
        return actual_damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        actual = target.take_damage(damage)
        print(f"{self.name}이 {target.name}에게 {actual} 만큼 공격했다!")

class Player(Character):
    def __init__(self, name):
        super().__init__(name, 100, 25, 5)
        self.level = 1
        self.exp = 0

    def gain_experience(self, exp):
        self.exp += exp
        if self.exp >= 50:
            self.level_up()

    def level_up(self):
        self.level += 1
        self.attack += 10
        self.defense += 5
        self.exp = 0
        print(f"레벨업! 현재 레벨: {self.level}")

class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        defen = random.randint(1, 5) * level
        super().__init__(name, hp, atk, defen)
        self.level = level

def battle(player, monster):
    print(f"\n{monster.name}과의 전투를 시작합니다.")
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
            print(f"{player.name}의 체력: {player.hp}")
        else:
            print(f"{monster.name} 처치!")

    if player.is_alive():
        player.gain_experience(monster.level * 20)
        print("전투 승리!")
        return True
    else:
        print("전투 패배..")
        return False

# 실행 예시
def main():
    player_name = input("플레이어 이름을 입력하세요: ")
    hero = Player(player_name)
    monsters = [Monster('슬라임', 1), Monster('고블린', 2), Monster('오크', 3)]

    for m in monsters:
        if not battle(hero, m):
            print("게임오버")
            break

main()
```
