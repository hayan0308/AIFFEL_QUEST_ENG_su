```python
class Square:          #클래스는 캐멀 케이스 사용
    def __init__(self):
        self.square = int(input("넓이 구할 사각형의 숫자를 고르시오.\n 1.평행사변형, 2.사다리꼴\n >>>"))

        if self.square == 1:
            print("평행사변형 함수는 par()입니다.")
            
        elif self.square == 2:                
            print("사다리꼴 함수는 trape()입니다.")
            
        else:
            print("숫자를 다시 입력해주세요.")


    def par(self):   # map(함수, 이터러블) 이터러블 : 반복할수있는 객체, ,map(int, )는 요소를 정수로 반환한다. 
        밑변, 높이 = map(int, input("평행사변형의 밑변과 높이를 입력해주세요.\n 예시: 밑변,높이\n : ").split(","))  # .split() : 괄호안에 입력한 값을 기준으로 분리
        계산식 = 밑변 * 높이 
        result = f"평행사변형의 넓이는 {계산식} 입니다."
        return result

    def trape(self):
        밑변, 윗변, 높이 = map(int, input("사다리꼴의 밑변과 윗변, 높이를 입력해주세요.\n 예시: 밑변,윗변,높이\n : ").split(","))
        계산식 = (밑변 + 윗변) * 높이 /2 
        result = f"사다리꼴의 넓이는 {계산식}입니다."
        return result


a = Square()    

a.par()
a.trape()
```
