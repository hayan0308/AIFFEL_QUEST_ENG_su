## Q1-2 회문 분류기

```python
# Q1-2 회문 분류기

# Q1-1에서 만든 판독기 함수를 이용하여 아래의 작업을 하는 회문 분류기 함수를 만들어 봅니다.



# targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

# print(palidrome_classifier(targets))
```

```python
# Q1-1에서 만든 판독기 함수를 이용
def is_palindrome(word):
    return word == word[::-1]


# 딕셔너리에 회문과 회문이 아닌 단어를 분류하여 저장
def palidrome_classifier(targets):
    result = {"회문O": [], "회문X": []}
    count_o = 0    # 문제 -'비율을 계산하여 반환' 여기에 필요

    
# 회문 분류 함수 필요
    
    # 단어 회문 체크 (def 안에 조건문)
    for word in targets:
        if is_palindrome(word):          
            result['회문O'].append(word)    # '회문O' 리스트에 추가
            count_o += 1                     # 리스트 안에 단어 개수 1 증가
        else:                                # 회문이 아니면?
            result['회문X'].append(word)    # '회문X' 리스트에 추가

    

# 분류가 완료 되면 저장한 딕셔너리와, 회문의 총 개수, 전체에서 회문의 비율을 계산하여 반환
    
   # 비율 계산 (회문 개수 / 전체 단어 개수)   
    ratio = count_o / len(targets)     # palindrome_classifier 함수에서 계산 (들여쓰기 위치 중요)
        
        
    return result, count_o, ratio       # return이 있어야만 함수를 호출한 곳에서 값을 받아와서 사용할 수 있다.


    
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우형우', '우향우']

print(palidrome_classifier(targets))
```

## Q2-2 성적 분석기

```python
# Q2-2 성적 분석기

#학생의 이름을 인자로 받아 아래의 예시와 같이 성적을 분석해주는 함수
# 학생 이름 입력했을때, socre_analysis("이름") 분석 출력


# 분석 함수 필요
# 학급 평균 값 필요, 평균 대비 점수 계산기 필요

# 전과목 평균 : "이름"의 평균 값, 학급평균(점수) 대비(내점수 -학급점수)
```

```python
scores = {
    "유원영": {"국어": 94, "영어": 91, "수학": 89},
    "설지민": {"국어": 88, "영어": 96, "수학": 90},
    "신윤아": {"국어": 86, "영어": 88, "수학": 77},
    "장윤아": {"국어": 98, "영어": 76, "수학": 92}
}

def score_analysis(student_name):
    if student_name not in scores:     # 1. 예외 처리 _이름이 장부에 있는지 확인
        print(f"'{student_name}' 학생의 데이터가 없습니다.")
        return

    # 2. 과목별 학급 총점 및 학생 수 계산을 위한 딕셔너리 준비
    subject_totals = {"국어": 0, "영어": 0, "수학": 0}
    num_students = len(scores)

    # 모든 학생의 점수를 돌며 과목별로 누적합을 구합니다.
    for name, student_scores in scores.items():
        for subject, score in student_scores.items():
            subject_totals[subject] += score

    # 과목별 학급 평균 계산
    class_sub_avg = {sub: total / num_students for sub, total in subject_totals.items()}
    
    # 전과목 학급 평균 (모든 과목 평균의 평균)
    class_total_avg = sum(class_sub_avg.values()) / len(class_sub_avg)

    # 3. 입력받은 학생의 데이터 가져오기
    my_scores = scores[student_name]
    my_total_avg = sum(my_scores.values()) / len(my_scores)

    # 4. 예시 양식에 맞춰 출력하기
    print(f"[{student_name} 성적 분석]")
    
    # 과목별 점수 및 학급 평균 대비 출력
    for subject, my_score in my_scores.items():
        sub_avg = class_sub_avg[subject]
        diff = my_score - sub_avg
        
        # 플러스일 때는 + 기호를 붙여주기 위해 기호 조건문 추가
        diff_str = f"+{diff:.2f}" if diff >= 0 else f"{diff:.2f}"
        print(f"{subject}: {my_score}점, 학급 평균({sub_avg:.2f}점) 대비 {diff_str}")

    print() # 한 줄 띄우기
    
    # 전과목 평균 출력
    total_diff = my_total_avg - class_total_avg
    total_diff_str = f"+{total_diff:.2f}" if total_diff >= 0 else f"{total_diff:.2f}"
    print(f"전과목 평균: {my_total_avg:.2f}점, 학급 평균({class_total_avg:.2f}점) 대비 {total_diff_str}")

# --- 실행 테스트 ---
score_analysis("설지민")
```
