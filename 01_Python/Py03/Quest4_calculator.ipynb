```python
import math

def calculator():
    print("=== 파이썬 간단한 계산기 ===")

    
    while True:
        # 첫 번째 정수 입력받기 (ValueError 처리 및 반복)
        while True:
            try:
                num1 = int(input("첫 번째 정수를 입력하세요: "))
                break  # 정상적으로 정수가 입력되면 루프 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")
                
        # 두 번째 정수 입력받기 (ValueError 처리 및 반복)
        while True:
            try:
                num2 = int(input("두 번째 정수를 입력하세요: "))
                break  # 정상적으로 정수가 입력되면 루프 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")
                
        #연산자 입력받기
        operator = input("연산자를 입력하세요 (+, -, *, /, **): ").strip()
        # strip() 앞뒤 공백 제거. 입력할때 공백 있어도 제거해서 정상처리함.
        
        # 계산 및 예외 처리 / 계산별 결과 처리 과정 필요
        try:
            if operator == '+':
                result = num1 + num2
                print(f"결과: {num1} + {num2} = {result}")
                
            elif operator == '-':
                result = num1 - num2
                print(f"결과: {num1} - {num2} = {result}")
                
            elif operator == '*':
                result = num1 * num2
                print(f"결과: {num1} * {num2} = {result}")
                
            elif operator == '/':
                # num2가 0이 되면 except ZeroDivisionError로 이동
                result = num1 / num2
                print(f"결과: {num1} / {num2} = {result}")

            elif operator == '**':
                # math 모듈을 사용한 제곱 연산 (결과는 float형으로 반환됨)
                result = math.pow(num1, num2)
                print(f"결과: {num1} ** {num2} = {result}")
                
            else:
                # 지원하지 않는 연산자 처리
                print("지원하지 않는 연산자입니다. (+, -, *, /, ** 중 하나를 입력해주세요)")
                
        
        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다.")
            
        
        
        print("-" * 40)

        
        # 계산기를 통해 계속 계산을 할 것인지
        continue_calc = input("계속 계산을 하시겠습니까? (y/n): ")
        if continue_calc == 'n':
            print("계산기를 종료합니다. 감사합니다!")
            break
        elif continue_calc == 'y':
            print("\n새로운 계산을 시작합니다.")



# 여러 개의 파이썬 파일을 엮어서 써야 할 땐, (if __name__ == "__main__":) 필요
calculator()
```
